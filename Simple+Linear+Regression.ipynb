### Simple Linear regression

Simple linear regression is a statistical method that allows us to summarize and study relationships between two continuous (quantitative) variables:
One variable, denoted x, is regarded as the predictor or independent variable.
The other variable, denoted y is regarded as dependent variable.

We will quickly jump in to coding part with real time dataset file name 'Salary_Data.csv'

```python
#Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importing the dataset
#While importing dataset please use forward slash '/' instead of backslash in source path to avoid Unicode error
dataset = pd.read_csv('C:/Users/rajsingh/Downloads/Salary_Data.csv')
#iloc is Location based indexing and square brackets indicates data from [Rows Indexs, coloumn Indexs] 
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
#Splitting the dataset into the Training set & Test set
#Please note  - 80:20 or 70:30 or accepted split ratio, However 50:50 is not a good ratio to split Training & Test set
#Random_state = 0 is to ensure random factors in algorithams do not impact the output result
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
#Fitting simple linear regression in to training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Predicting test set results
y_pred = regressor.predict(X_test)
```

```python
#Visualising the Training set results
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience(Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualising the test set results
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience(Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

Finally, we can plot the predictions as a line and compare it to the original dataset and ensure our test set is as linear as train set!!!
